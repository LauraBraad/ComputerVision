```python
import json
import pandas as pd
```

```python
path = './coco-labels-paper.txt'
labels = pd.read_csv(path)
labels['category_id'] = labels.index
```

```python
pred = open('./yolov7_predictions.json')
pred = pd.read_json(pred)
```

```python
pred['category_id_new'] = pred['category_id'] - 1
```

```python
pred_mean = pred.groupby("category_id_new").agg({'score': ['count', 'mean']})
```

```python
merged_df = pd.merge(pred_mean, labels, left_on = 'category_id_new', right_on = 'category_id' , how='inner')

# Display the merged DataFrame
merged_df
```
